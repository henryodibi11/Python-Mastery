## Solution 1: Basic Assertions

Comprehensive tests for string utilities with edge cases.

```python
%%writefile test_strings_solution.py
"""Solution: Basic assertion tests for string utilities."""
import pytest

def reverse_string(s: str) -> str:
    return s[::-1]

def is_palindrome(s: str) -> bool:
    s_clean = s.lower().replace(" ", "")
    return s_clean == s_clean[::-1]

def count_vowels(s: str) -> int:
    return sum(1 for c in s.lower() if c in 'aeiou')

# reverse_string tests
def test_reverse_simple_string():
    assert reverse_string("hello") == "olleh"

def test_reverse_empty_string():
    assert reverse_string("") == ""

def test_reverse_single_char():
    assert reverse_string("a") == "a"

def test_reverse_with_spaces():
    assert reverse_string("hello world") == "dlrow olleh"

def test_reverse_with_numbers():
    assert reverse_string("abc123") == "321cba"

# is_palindrome tests
def test_palindrome_simple():
    assert is_palindrome("racecar") == True

def test_palindrome_case_insensitive():
    assert is_palindrome("RaceCar") == True

def test_palindrome_with_spaces():
    assert is_palindrome("race car") == True

def test_not_palindrome():
    assert is_palindrome("hello") == False

def test_palindrome_empty():
    assert is_palindrome("") == True

def test_palindrome_single_char():
    assert is_palindrome("a") == True

# count_vowels tests
def test_count_vowels_all_vowels():
    assert count_vowels("aeiou") == 5

def test_count_vowels_mixed():
    assert count_vowels("hello world") == 3

def test_count_vowels_none():
    assert count_vowels("xyz") == 0

def test_count_vowels_empty():
    assert count_vowels("") == 0

def test_count_vowels_case_insensitive():
    assert count_vowels("AEIOU") == 5

def test_count_vowels_with_numbers():
    assert count_vowels("test123") == 1
```

```python
!pytest test_strings_solution.py -v
```

**Key Learning**: Always test edge cases - empty strings, single characters, special characters, case sensitivity.

## Solution 2: Fixtures

Reusable fixtures for shopping cart tests.

```python
%%writefile test_cart_solution.py
"""Solution: Fixture-based tests for ShoppingCart."""
import pytest

class ShoppingCart:
    def __init__(self):
        self.items = []
    
    def add_item(self, name: str, price: float, quantity: int = 1):
        self.items.append({"name": name, "price": price, "quantity": quantity})
    
    def total(self) -> float:
        return sum(item["price"] * item["quantity"] for item in self.items)
    
    def item_count(self) -> int:
        return sum(item["quantity"] for item in self.items)
    
    def clear(self):
        self.items.clear()

# Fixtures
@pytest.fixture
def empty_cart():
    """Provide empty shopping cart."""
    return ShoppingCart()

@pytest.fixture
def cart_with_items():
    """Provide cart pre-populated with items."""
    cart = ShoppingCart()
    cart.add_item("Apple", 1.50, 3)
    cart.add_item("Banana", 0.75, 2)
    cart.add_item("Orange", 2.00, 1)
    return cart

@pytest.fixture(scope="module")
def product_catalog():
    """Product catalog - created once per module."""
    return {
        "Apple": 1.50,
        "Banana": 0.75,
        "Orange": 2.00,
        "Milk": 3.50,
    }

# Tests using empty_cart
def test_empty_cart_total(empty_cart):
    assert empty_cart.total() == 0

def test_empty_cart_item_count(empty_cart):
    assert empty_cart.item_count() == 0

def test_add_item_to_empty_cart(empty_cart):
    empty_cart.add_item("Apple", 1.50)
    assert empty_cart.item_count() == 1
    assert empty_cart.total() == 1.50

# Tests using cart_with_items
def test_cart_total(cart_with_items):
    # (1.50 * 3) + (0.75 * 2) + (2.00 * 1) = 4.50 + 1.50 + 2.00 = 8.00
    assert cart_with_items.total() == 8.00

def test_cart_item_count(cart_with_items):
    # 3 + 2 + 1 = 6 items
    assert cart_with_items.item_count() == 6

def test_add_item_to_existing_cart(cart_with_items):
    cart_with_items.add_item("Milk", 3.50, 1)
    assert cart_with_items.item_count() == 7
    assert cart_with_items.total() == 11.50

def test_clear_cart(cart_with_items):
    cart_with_items.clear()
    assert cart_with_items.item_count() == 0
    assert cart_with_items.total() == 0

# Test using module-scoped fixture
def test_product_catalog_lookup(product_catalog):
    assert product_catalog["Apple"] == 1.50
    assert "Milk" in product_catalog
```

```python
!pytest test_cart_solution.py -v
```

**Key Learning**: 
- Use fixtures to avoid code duplication
- Function scope (default) creates fresh instance per test
- Module scope shares instance across tests (for read-only data)

## Solution 3: Parametrize

Data-driven tests for calculator with named IDs.

```python
%%writefile test_calculator_solution.py
"""Solution: Parametrized tests for calculator."""
import pytest

def calculate(operation: str, a: float, b: float) -> float:
    if operation == "add":
        return a + b
    elif operation == "subtract":
        return a - b
    elif operation == "multiply":
        return a * b
    elif operation == "divide":
        if b == 0:
            raise ValueError("Cannot divide by zero")
        return a / b
    else:
        raise ValueError(f"Unknown operation: {operation}")

# Parametrized tests for valid operations
@pytest.mark.parametrize("operation,a,b,expected", [
    # Addition
    ("add", 2, 3, 5),
    ("add", -1, 1, 0),
    ("add", 0, 0, 0),
    ("add", 1.5, 2.5, 4.0),
    ("add", -5, -3, -8),
    
    # Subtraction
    ("subtract", 5, 3, 2),
    ("subtract", 10, 10, 0),
    ("subtract", 0, 5, -5),
    ("subtract", -3, -5, 2),
    ("subtract", 7.5, 2.5, 5.0),
    
    # Multiplication
    ("multiply", 3, 4, 12),
    ("multiply", -2, 5, -10),
    ("multiply", 0, 100, 0),
    ("multiply", 1.5, 2, 3.0),
    ("multiply", -3, -4, 12),
    
    # Division
    ("divide", 10, 2, 5),
    ("divide", 9, 3, 3),
    ("divide", -10, 2, -5),
    ("divide", 7, 2, 3.5),
    ("divide", 0, 5, 0),
], ids=[
    # Addition IDs
    "add_positive", "add_negative_positive", "add_zeros", "add_floats", "add_negatives",
    # Subtraction IDs
    "sub_positive", "sub_equal", "sub_from_zero", "sub_negatives", "sub_floats",
    # Multiplication IDs
    "mul_positive", "mul_negative", "mul_by_zero", "mul_floats", "mul_negatives",
    # Division IDs
    "div_positive", "div_exact", "div_negative", "div_float_result", "div_zero",
])
def test_calculate_valid_operations(operation, a, b, expected):
    """Test all valid calculator operations."""
    result = calculate(operation, a, b)
    assert result == pytest.approx(expected)

# Parametrized tests for error cases
@pytest.mark.parametrize("operation,a,b,expected_error", [
    ("divide", 10, 0, "Cannot divide by zero"),
    ("divide", -5, 0, "Cannot divide by zero"),
    ("invalid", 1, 2, "Unknown operation: invalid"),
    ("power", 2, 3, "Unknown operation: power"),
], ids=["div_by_zero_positive", "div_by_zero_negative", "invalid_op", "unknown_op"])
def test_calculate_errors(operation, a, b, expected_error):
    """Test error handling."""
    with pytest.raises(ValueError, match=expected_error):
        calculate(operation, a, b)

# Alternative: Separate parametrized test per operation
@pytest.mark.parametrize("a,b,expected", [
    (2, 3, 5),
    (10, 15, 25),
    (-5, 5, 0),
    (0, 0, 0),
    (1.5, 2.5, 4.0),
])
def test_addition_focused(a, b, expected):
    """Focused tests for addition only."""
    assert calculate("add", a, b) == pytest.approx(expected)
```

```python
!pytest test_calculator_solution.py -v
```

**Key Learning**:
- Parametrize reduces code duplication from 20+ functions to 2
- Named IDs make test output readable
- Separate error tests from valid operation tests
- Use `pytest.approx()` for float comparisons

## Solution 4: Mocking

Comprehensive mocking for external dependencies.

```python
%%writefile test_weather_solution.py
"""Solution: Mocking external dependencies."""
import pytest
from unittest.mock import Mock, patch
import os

# Code under test
import requests

class WeatherService:
    def __init__(self):
        self.api_key = os.getenv("WEATHER_API_KEY")
        self.base_url = "https://api.weather.com"
    
    def get_temperature(self, city: str) -> float:
        if not self.api_key:
            raise ValueError("API key not configured")
        
        url = f"{self.base_url}/current?city={city}&key={self.api_key}"
        response = requests.get(url)
        response.raise_for_status()
        
        data = response.json()
        return data["temperature"]
    
    def is_cold(self, city: str) -> bool:
        temp = self.get_temperature(city)
        return temp < 10

# Tests
def test_get_temperature_success(monkeypatch):
    """Test successful API call."""
    # Mock environment variable
    monkeypatch.setenv("WEATHER_API_KEY", "test-key-123")
    
    # Mock requests.get
    mock_response = Mock()
    mock_response.status_code = 200
    mock_response.json.return_value = {"temperature": 22.5}
    
    def mock_get(url):
        return mock_response
    
    monkeypatch.setattr(requests, "get", mock_get)
    
    # Test
    service = WeatherService()
    temp = service.get_temperature("London")
    assert temp == 22.5

def test_missing_api_key(monkeypatch):
    """Test error when API key not configured."""
    # Ensure WEATHER_API_KEY is not set
    monkeypatch.delenv("WEATHER_API_KEY", raising=False)
    
    service = WeatherService()
    with pytest.raises(ValueError, match="API key not configured"):
        service.get_temperature("London")

def test_api_error(monkeypatch):
    """Test handling of API errors."""
    monkeypatch.setenv("WEATHER_API_KEY", "test-key")
    
    # Mock failed request
    mock_response = Mock()
    mock_response.raise_for_status.side_effect = requests.HTTPError("404 Not Found")
    
    monkeypatch.setattr(requests, "get", lambda url: mock_response)
    
    service = WeatherService()
    with pytest.raises(requests.HTTPError):
        service.get_temperature("InvalidCity")

@pytest.mark.parametrize("temp,expected_cold", [
    (5.0, True),    # Below 10
    (9.9, True),    # Just below 10
    (10.0, False),  # Exactly 10
    (15.0, False),  # Above 10
    (-5.0, True),   # Negative temp
])
def test_is_cold(monkeypatch, temp, expected_cold):
    """Test is_cold with various temperatures."""
    monkeypatch.setenv("WEATHER_API_KEY", "test-key")
    
    # Mock get_temperature to return specific temp
    mock_response = Mock()
    mock_response.json.return_value = {"temperature": temp}
    monkeypatch.setattr(requests, "get", lambda url: mock_response)
    
    service = WeatherService()
    assert service.is_cold("TestCity") == expected_cold

# Alternative: Using pytest-mock plugin
def test_with_pytest_mock(mocker):
    """Test using pytest-mock plugin (cleaner syntax)."""
    # Mock environment
    mocker.patch.dict(os.environ, {"WEATHER_API_KEY": "test-key"})
    
    # Mock requests.get
    mock_get = mocker.patch("requests.get")
    mock_get.return_value.json.return_value = {"temperature": 18.0}
    
    service = WeatherService()
    temp = service.get_temperature("Paris")
    
    assert temp == 18.0
    mock_get.assert_called_once()  # Verify mock was called
```

**Key Learning**:
- Use `monkeypatch` to replace functions, env vars, attributes
- Mock external dependencies to avoid real API calls
- Test both success and error paths
- `pytest-mock` provides cleaner syntax than unittest.mock
